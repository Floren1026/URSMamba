```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 读取彩色图像
# image = cv2.imread('/home/user-yc/MRSIS/result_analysis/result/cover/cover_img0000.png')
image = cv2.imread('/home/user-yc/MRSIS/result_analysis/result/steg/1/steg_img0000.png')


# 创建保存目录（可选）
save_dir = 'histogram_plots'
os.makedirs(save_dir, exist_ok=True)


# 将图像转为灰度图（用于整体亮度直方图）
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 计算直方图
hist_r = cv2.calcHist([image], [2], None, [256], [0, 256]).flatten()  # R通道
hist_g = cv2.calcHist([image], [1], None, [256], [0, 256]).flatten()  # G通道
hist_b = cv2.calcHist([image], [0], None, [256], [0, 256]).flatten()  # B通道
hist_gray = cv2.calcHist([gray], [0], None, [256], [0, 256]).flatten()  # 灰度图

x = np.arange(256)

# 显示图像和直方图
plt.figure(figsize=(6, 6))
plt.fill_between(x, hist_r, color='red', alpha=0.7)
plt.tick_params(axis='both', labelsize=18)  # 坐标轴刻度字体
plt.savefig(f'{save_dir}/red_histogram.png', dpi=600, bbox_inches='tight')
plt.close()

# G通道直方图
plt.figure(figsize=(6, 6))
plt.fill_between(x, hist_g, color='green', alpha=0.7)
plt.tick_params(axis='both', labelsize=18)  # 坐标轴刻度字体
plt.savefig(f'{save_dir}/green_histogram.png', dpi=600, bbox_inches='tight')
plt.close()

# B通道直方图
plt.figure(figsize=(6, 6))
plt.fill_between(x, hist_b, color='blue', alpha=0.7)
plt.tick_params(axis='both', labelsize=18)  # 坐标轴刻度字体
plt.savefig(f'{save_dir}/blue_histogram.png', dpi=600, bbox_inches='tight')
plt.close()

# 灰度直方图
plt.figure(figsize=(6, 6))
plt.fill_between(x, hist_gray, color='grey', alpha=0.7)
plt.tick_params(axis='both', labelsize=18)  # 坐标轴刻度字体
plt.savefig(f'{save_dir}/combined_histogram.png', dpi=600, bbox_inches='tight')
plt.close()

plt.tight_layout()
plt.show()
```
